```python
import pandas as pd
```

```python
all_data = pd.read_csv('all_data.csv')
```

```python
all_data.head()
```

```python
close_columns = ['date']+[col for col in all_data.columns if 'close' in col] 
close_columns
```

```python
close_data = all_data[close_columns]

#rename date to ts_event
close_data = close_data.rename(columns={'date':'ts_event'})
```

```python
close_data.head()

close_data['month'] = close_data.ts_event.dt.month
```

```python
close_data.to_csv('./dataset/close_data.csv', index=False)
```
